```python
data = pd.read_excel('data/dataset.xlsx', index=False, index_col=None, header=0, sheet_name='Sheet1', skiprows=0)

data = shuffle(data)

data.head()
# data.tail()
```

```python
print("Data Features:", data.columns)
print("Data Dimension:", data.ndim)
print("Data Shape:", data.shape)
print("Data Size:", data.size)
```

```python
celsius = data['Celsius'].values
fahrenheit = data['Fahrenheit'].values

for idx, c in enumerate(celsius[0:10]):
  print("{}. {} Celsius ==> {} Fahrenheit".format(idx+1, c, round(fahrenheit[idx],3)))
```

### Create the Neural Network Model

```python
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
```

```python
model = tf.keras.Sequential([l0])
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celsius, fahrenheit, epochs=1000, verbose=True)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
print("These are the layer variables: {}".format(l0.get_weights()))
```

```python
# print(l0.get_weights())
print("Fahrenheit = Celsius*{} + {}".format(float(l0.get_weights()[0]), float(l0.get_weights()[1])))
```

```python
# print(len(history.history['loss']))
```

```python
print("Celsius to Fahrenheit with using Neural Network")
print("100 Celsius equal to Fahrenheit:", model.predict([100]))
```

## <center>Create Function: Cel2Fah</center>

```python
def Cel2Fah(C):
    F = C * 1.8 + 32
    return F
```

```python
print("Celsius to Fahrenheit:", Cel2Fah(100))
```
